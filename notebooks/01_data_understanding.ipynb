Step 1

Get failures for machine 1.

```python
failure_time = pd.Timestamp(
    "2015-01-05 06:00:00"
)
```

Step 3

Look 72 hours before failure.

```python
machine1 = telemetry[
    telemetry["machineID"] == 1
]

window = machine1[
    (machine1["datetime"] >= failure_time - pd.Timedelta(hours=72))
    &
    (machine1["datetime"] <= failure_time)
]
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    window["datetime"],
    window["vibration"]
)

plt.axvline(
    failure_time,
    color="red",
    linestyle="--",
    label="Failure"
)

plt.legend()

plt.title(
    "Vibration Before Failure"
)

plt.show()
```

Vibration (Most Interesting)

The vibration plot shows something important:

Early part of the window: mostly between 30–45
Later part (closer to failure): mostly between 48–60
Multiple spikes above 55

This suggests:

Vibration appears to increase as failure approaches.
